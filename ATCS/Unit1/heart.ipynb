## 1. Heart Attack Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import statsmodels as sm
```

```python
df = pd.read_csv("heart.csv")
df.describe()
```

```python
# pie chart to show the chance of being high risk vs. low risk for males and females
fig, ax = plt.subplots(1, 2, figsize = (10,5))
risks = ['Low risk', 'High risk']
ax[0].pie(df.groupby(['sex', 'output']).size()[0], labels = risks, autopct='%1.1f%%')
ax[0].set_title('Male')
ax[1].pie(df.groupby(['sex', 'output']).size()[1], labels = risks, autopct='%1.1f%%')
ax[1].set_title('Female')
fig.suptitle('Sex vs. risk of heart attack')
fig.subplots_adjust(top = .9)
plt.show()
```

```python
# pie chart to show the chance of being high risk vs. low risk for different chest pain types
fig, ax = plt.subplots(1, 4, figsize = (15,5))
risks = ['Low risk', 'High risk']
ax[0].pie(df.groupby(['cp', 'output']).size()[0], labels = risks, autopct='%1.1f%%')
ax[0].set_title('Typical Angina')
ax[1].pie(df.groupby(['cp', 'output']).size()[1], labels = risks, autopct='%1.1f%%')
ax[1].set_title('Atypical Angina')
ax[2].pie(df.groupby(['cp', 'output']).size()[2], labels = risks, autopct='%1.1f%%')
ax[2].set_title('Non-anginal pain')
ax[3].pie(df.groupby(['cp', 'output']).size()[3], labels = risks, autopct='%1.1f%%')
ax[3].set_title('Asymptomatic')
fig.suptitle('Type of chest pain vs. risk of heart attack')
fig.subplots_adjust(top = 1)
plt.show()
```

```python
# pie chart to show the chance of being high risk vs. low risk for subjects whose fasting blood sugar was above or below 120 mg/dl
fig, ax = plt.subplots(1, 2, figsize = (10,5))
risks = ['Low risk', 'High risk']
ax[0].pie(df.groupby(['fbs', 'output']).size()[0], labels = risks, autopct='%1.1f%%')
ax[0].set_title('fbs < 120 mg/dl')
ax[1].pie(df.groupby(['fbs', 'output']).size()[1], labels = risks, autopct='%1.1f%%')
ax[1].set_title('fbs > 120 mg/dl')
fig.suptitle('Fasting blood sugar vs. risk of heart attack')
fig.subplots_adjust(top = .9)
plt.show()
```

```python
# pie chart to show the chance of being high risk vs. low risk for different electrocardiographic results
fig, ax = plt.subplots(1, 3, figsize = (15,5))
risks = ['Low risk', 'High risk']
ax[0].pie(df.groupby(['restecg', 'output']).size()[0], labels = risks, autopct='%1.1f%%')
ax[0].set_title('Normal')
ax[1].pie(df.groupby(['restecg', 'output']).size()[1], labels = risks, autopct='%1.1f%%')
ax[1].set_title('ST-T wave abnormality')
ax[2].pie(df.groupby(['restecg', 'output']).size()[2], labels = risks, autopct='%1.1f%%')
ax[2].set_title('Probable or definite left ventricular hypertrophy')
fig.suptitle('Resting electrocardiographic results vs. risk of heart attack')
fig.subplots_adjust(top = .9)
plt.show()
print(df.groupby(['restecg', 'output']).size())
```

```python
# pie chart to show the chance of being high risk vs. low risk based on whether subjects' exercise induced angina or not
fig, ax = plt.subplots(1, 2, figsize = (10,5))
risks = ['Low risk', 'High risk']
ax[0].pie(df.groupby(['exng', 'output']).size()[0], labels = risks, autopct='%1.1f%%')
ax[0].set_title('Exercise did not induce angina')
ax[1].pie(df.groupby(['exng', 'output']).size()[1], labels = risks, autopct='%1.1f%%')
ax[1].set_title('Exercise induced angina')
fig.suptitle('Whether exercise induced angina or not vs. risk of heart attack')
fig.subplots_adjust(top = .9)
plt.show()
```

```python
# pie chart to show the chance of being high risk vs. low risk based the slope of the peak exercise ST segment
fig, ax = plt.subplots(1, 3, figsize = (10,5))
risks = ['Low risk', 'High risk']
ax[0].pie(df.groupby(['slp', 'output']).size()[0], labels = risks, autopct='%1.1f%%')
ax[0].set_title('Upsloping')
ax[1].pie(df.groupby(['slp', 'output']).size()[1], labels = risks, autopct='%1.1f%%')
ax[1].set_title('Flat')
ax[2].pie(df.groupby(['slp', 'output']).size()[2], labels = risks, autopct='%1.1f%%')
ax[2].set_title('Downsloping')
fig.suptitle('Slope of the peak exercise ST segment vs. risk of heart attack')
fig.subplots_adjust(top = 1.2)
plt.show()
```

```python
# pie chart to show the chance of being high risk vs. low risk based on the number of major vessels colored by fluoroscopy
fig, ax = plt.subplots(1, 5, figsize = (20,5))
risks = ['Low risk', 'High risk']
ax[0].pie(df.groupby(['caa', 'output']).size()[0], labels = risks, autopct='%1.1f%%')
ax[0].set_title('0 major vessels')
ax[1].pie(df.groupby(['caa', 'output']).size()[1], labels = risks, autopct='%1.1f%%')
ax[1].set_title('1 major vessel')
ax[2].pie(df.groupby(['caa', 'output']).size()[2], labels = risks, autopct='%1.1f%%')
ax[2].set_title('2 major vessels')
ax[3].pie(df.groupby(['caa', 'output']).size()[3], labels = risks, autopct='%1.1f%%')
ax[3].set_title('3 major vessels')
ax[4].pie(df.groupby(['caa', 'output']).size()[4], labels = risks, autopct='%1.1f%%')
ax[4].set_title('4 major vessels')
fig.suptitle('Number of major vessels colored by fluoroscopy vs. risk of heart attack')
fig.subplots_adjust(top = .9)
plt.show()
```

```python
# logistic regression to estimate the probability of having a higher chance of a heart attack with 95% confidence interval
sns.regplot(x='age', y='output', data=df, logistic=True, y_jitter = 0.03).set(title = 'Age vs. risk of a heart attack')
plt.show()
sns.regplot(x='trtbps', y='output', data=df, logistic=True, y_jitter = 0.03).set(title = 'Resting blood pressure (mmHg) vs. risk of a heart attack')
plt.show()
sns.regplot(x='chol', y='output', data=df, logistic=True, y_jitter = 0.03).set(title = 'Cholesterol level (mg/dl) vs. risk of a heart attack')
plt.show()
sns.regplot(x='thalachh', y='output', data=df, logistic=True, y_jitter = 0.03).set(title = 'Maximum heart rate achieved vs. risk of a heart attack')
plt.show()
sns.regplot(x='oldpeak', y='output', data=df, logistic=True, y_jitter = 0.03).set(title = 'ST depression induced by exercise relative to rest vs. risk of a heart attack')
plt.show()
```

```python
# visualizing correlations between quantitative features
plt = sns.pairplot(df, x_vars = ['age', 'trtbps', 'chol', 'thalachh', 'oldpeak'], y_vars = ['age', 'trtbps', 'chol', 'thalachh', 'oldpeak'])
plt.fig.suptitle('Pair plot for quantitative features')
plt.fig.subplots_adjust(top = 0.95)
```

```python
sns.set(rc={'figure.figsize':(15,15)})
correlationMatrix = df.corr()
fig = sns.heatmap(correlationMatrix, square = True, annot = True, center = 0, cmap = 'seismic').set(title='Correlation between features')
```
